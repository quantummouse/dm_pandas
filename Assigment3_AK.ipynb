I've created visualizaitons to show differences in the state reimbursements for medicaid, which drugs are usually reimbursed as well the number of prescriptions written for the drugs. I wanted to see if the number of prescirptions and units reimburse would have a linear relationship, but they don't suggesting that some drugs are prescribed but might not be reimbursed. The charts also show top 10 states for reimbursements and top 10 drugs for perscriptions and reimbursements. Ibuprofin and Hydrocodon have an interesting releationships as there is an opiod epidemic and both are pain killers. 

```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
drugsutil = pd.read_csv("/content/drive/My Drive/Colab Notebooks/State_Drug_Utilization_Data_2010.csv")
drugsutil.head(4)
```

```
g = sns.barplot( "Total Amount Reimbursed", "State", data=drugsutil, order=pd.value_counts(drugsutil['State']).iloc[:10].index)

plt.show()
```

```
h = sns.barplot( "Medicaid Amount Reimbursed", "State", data=drugsutil, order=pd.value_counts(drugsutil['State']).iloc[:10].index)
plt.show()
```

```
g = sns.barplot( "Total Amount Reimbursed", "Product Name", data=drugsutil, order=pd.value_counts(drugsutil['Product Name']).iloc[:15].index)
```

```
 k= sns.barplot( "Medicaid Amount Reimbursed", "Product Name", data=drugsutil, order=pd.value_counts(drugsutil['Product Name']).iloc[:15].index)
```

```
k= sns.barplot( "Non Medicaid Amount Reimbursed", "Product Name", data=drugsutil, order=pd.value_counts(drugsutil['Product Name']).iloc[:15].index)

```

```

m= sns.barplot( "Number of Prescriptions", "Product Name", data=drugsutil, order=pd.value_counts(drugsutil['Product Name']).iloc[:15].index)

```

```
plt.figure(figsize=(20,20))
p= sns.catplot("Product Name", "Number of Prescriptions", data = drugsutil, order=pd.value_counts(drugsutil['Product Name']).iloc[:10].index )
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)
```
